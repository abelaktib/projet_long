```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
```

```python
## path file result
str_filename="bigdatapr8.csv"
```

# Open the file and check header

```python
fh=open(str_filename)
csv_reader=csv.reader(fh)
csv_header=next(csv_reader)
csv_header
```

# Visualize the file

```python
df_sig=pd.read_csv(str_filename,names=csv_header,header=0)
df_sig
```

# LOSS & ACCURACY

```python
plt.plot(df_sig['ACCURACY'])
plt.plot(df_sig['VAL_ACCURACY'])
plt.plot(df_sig['LOSS'])
plt.plot(df_sig['VAL_LOSS'])
plt.legend(['ACCURACY','VAL_ACCURACY','LOSS','VAL_LOSS'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

# LOSS ONLY

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=False)
ax1.plot(df_sig['LOSS'])
ax2.plot(df_sig['VAL_LOSS'])
ax1.legend(['Train','VALIDATION'])
ax2.legend(['VALIDATION'])
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.tight_layout()

plt.show()
```

# Others metrics

```python
plt.plot(df_sig['ROC'])
plt.plot(df_sig['VAL_ROC'])
plt.plot(df_sig['PR'])
plt.plot(df_sig['VAL_PR'])


plt.plot(df_sig['PRECISION'])
plt.plot(df_sig['VAL_PRECISION'])
plt.plot(df_sig['RECALL'])
plt.plot(df_sig['VAL_RECALL'])
plt.plot(df_sig['BIN_ACC'])
plt.plot(df_sig['VAL_BIN_ACC'])


plt.legend(['ROC','VAL_ROC','PR','VAL_PR','PRECISION','VAL_PRECISION','RECALL','VAL_RECALL','BIN_ACC','VAL_BIN_ACC'])
plt.xlabel('Epoch')
plt.ylim(0,1)
plt.show()
```

```python
plt.plot(df_sig['ROC'])
plt.plot(df_sig['PR'])


plt.plot(df_sig['PRECISION'])
plt.plot(df_sig['RECALL'])
plt.plot(df_sig['BIN_ACC'])


plt.legend(['AUROC','AUPR','PRECISION','RECALL','BIN_ACC'])
plt.xlabel('Epoch')
plt.ylim(0,1)
plt.show()
```

```python
plt.plot(df_sig['VAL_ROC'])
plt.plot(df_sig['VAL_PR'])

plt.plot(df_sig['VAL_PRECISION'])
plt.plot(df_sig['VAL_RECALL'])
plt.plot(df_sig['VAL_BIN_ACC'])

plt.legend(['VAL_AUROC','VAL_AUPR','VAL_PRECISION','VAL_RECALL','VAL_BIN_ACC'])
plt.xlabel('Epoch')
plt.ylim(0,1)
plt.show()
```
